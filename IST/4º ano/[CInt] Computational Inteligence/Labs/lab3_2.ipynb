## Use  the  file  DCOILBRENTEUv2.csv  as  your  data  file.  Read  the  file  and  plot  it  using pandas  and  matplotlib.  Make  sure  the  program  can  open  the  file  correctly  and  plot  it correctly.  

```python
from unicodedata import decimal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import *

#read csv file
df = pd.read_csv("DCOILBRENTEUv2.csv", decimal='.',sep=',')

#convert date to datetime
df['DATE'] = df['DATE'].astype(np.datetime64)
 
#Plot the data
plt.plot(df['DATE'], df['DCOILBRENTEU'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## Plot a histogram where in the x axis you plot the variation from the previous day (Close from the previous day to Close from the present day). 

```python
#Plot a histogram where in the x axis you plot the variation from the previous day (Close from the previous day to Close from the present day).
#The y axis is the number of days that the variation was in that range.
#The histogram should have 20 bins.
#The histogram should have a title and the x and y axis should be labeled.
#The histogram should be saved as a png file.

#calculate the variation from the previous day
df['variation'] = df['DCOILBRENTEU'].diff()

#plot the histogram
plt.hist(df['variation'], bins=20)
plt.xlabel('Variation')
plt.ylabel('Number of days')
plt.title('Histogram of variation')
df.to_csv('DCOILBRENTEUv2_changed.csv', index=True)
plt.show()
```
