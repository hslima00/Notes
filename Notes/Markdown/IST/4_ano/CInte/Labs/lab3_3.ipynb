# Lab 3 Exercise 3 - Scatter Plot

## Use  the  file  DCOILBRENTEUv2.csv  and  DCOILWTICOv2.csv  as  your  data  files. 

```python
from unicodedata import decimal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import *

# Read the data
breant_df = pd.read_csv("DCOILBRENTEUv2.csv", decimal='.',sep=',')
wtico_df = pd.read_csv('DCOILWTICOv2.csv', decimal='.',sep=',')

#Data to datetime
breant_df['DATE'] = breant_df['DATE'].astype(np.datetime64)
wtico_df['DATE'] = wtico_df['DATE'].astype(np.datetime64)


```

## Brent is the price of oil in the UK in dollars and WTI (West Texas Intermediate) is the price  of  oil  in  Texas,  USA  also  in  dollars.  They  have  similar  values,  but  some  time differ a little because of transportation and refining issues. 

## Read the files and plot then in  a  single  figure  using  pandas  and  matplotlib.  Make  sure  the  program  is  able  to  open the file correctly and plot it correctly.  

```python
#Plot the data in the same figure

plt.plot(breant_df['DATE'],breant_df['DCOILBRENTEU'], label='Brent')
plt.plot(wtico_df['DATE'],wtico_df['DCOILWTICO'], label='WTI')
plt.legend()
plt.show()
```

## Now  make  a  Scatter  plot  with  the  value  oil  of  the  different  regions  and  see  how  they behave.  


## As an extra, added the trendline

```python
#make the dataframe size equal 
if len(breant_df) > len(wtico_df):
    breant_df = breant_df[:len(wtico_df)]
else:
    wtico_df = wtico_df[:len(breant_df)]

# scatter plot the data
plt.scatter(breant_df['DCOILBRENTEU'],'b',wtico_df['DCOILWTICO'], 'r')
plt.xlabel('Brent')
plt.ylabel('WTI')

# calculate the trendline
z = np.polyfit(breant_df['DCOILBRENTEU'],wtico_df['DCOILWTICO'], 1)

# plot the trendline
p = np.poly1d(z)
plt.plot(breant_df['DCOILBRENTEU'],p(breant_df['DCOILBRENTEU']),"r--")
plt.show()
```
